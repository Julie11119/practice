```python
import pandas as pd
```

```python
dataset = pd.read_csv('car_price.csv')
```

```python
dataset_copy = dataset.copy()
```

```python
dataset_copy
```

```python
dataset_copy.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
columns_to_convert = list(dataset_copy.select_dtypes(include='object').columns)
columns_to_convert.remove('CarName')
for i in columns_to_convert:
    dataset_copy[i + 'copy']= le.fit_transform(dataset_copy[i])
dataset_copy_new = dataset_copy.drop(columns_to_convert,axis=1)
dataset_copy_new = dataset_copy_new.drop(columns ='CarNamecopy')
```

```python
dataset_copy_new
```

```python
import numpy as np
cor_matrix = dataset_copy_new.corr().abs()
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]
dataset_copy_new1 = dataset_copy_new.drop(dataset_copy_new[to_drop], axis=1)
dataset_copy_new1
```

```python
co = dataset_copy_new1.columns
dataset_copy_new1 = dataset_copy_new1.loc[:,['car_ID', 'CarName','symboling', 'wheelbase', 'carlength', 'carwidth',
       'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'price',
       'aspirationcopy', 'doornumbercopy', 'carbodycopy', 'drivewheelcopy',
       'enginelocationcopy', 'enginetypecopy', 'cylindernumbercopy',
       'fuelsystemcopy']]
dataset_copy_new1
```

```python
from sklearn.preprocessing import MinMaxScaler
data = dataset_copy_new1.iloc[:,2:].values
labels = dataset_copy_new1.iloc[:,:2].values
mm = MinMaxScaler()
data_normalized = mm.fit_transform(data)
new_data0 = pd.DataFrame(data_normalized ,columns=dataset_copy_new1.columns[2:])
new_data1 = pd.DataFrame(labels ,columns=dataset_copy_new1.columns[:2])
new_data = pd.concat([new_data1, new_data0], axis = 1)
new_data
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
predict_y = kmeans.fit_predict(new_data.iloc[:,2:])
final_data = pd.DataFrame(predict_y,columns=['result'])
final_result = pd.concat([dataset_copy_new1, final_data], axis = 1)
final_result
```
